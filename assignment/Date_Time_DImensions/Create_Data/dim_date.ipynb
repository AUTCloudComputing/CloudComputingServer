# Create DIM Date Table

This notebook is populated a date dimension for a data warehouse that requires time analysis and intelligence.

```python
import pandas as pd
import numpy as np
import datetime
```

```python
# Create date list
current_date = datetime.date(2000,1,1)
end_date = datetime.date(2050, 12, 31)
date_list = []
```

```python
while current_date < end_date:
    current_date += datetime.timedelta(days=1)
    date_list.append(current_date)
```

```python
date_list
```

```python
date_df = pd.DataFrame({'date': date_list})
date_df.describe()
```

```python
# dim_date_key INT PRIMARY KEY NOT NULL,
date_df['dim_date_key'] = date_df['date'].apply(lambda x: x.strftime("%Y%m%d"))

```

```python
    #  calendar_date DATETIME NOT NULL,
date_df['calendar_date'] = date_df['date']
```

```python
#     cal_year INT NOT NULL, 
#     cal_week_in_year INT NOT NULL,
#      cal_day_in_week_no INT NOT NULL
date_df['cal_year'], date_df['week_in_year'], date_df['day_in_week'] = zip(*date_df['date'].apply(lambda x: x.isocalendar()))
```

```python
#     cal_day_in_year INT NOT NULL,
date_df['cal_day_in_year'] = date_df['date'].apply(lambda x: x.timetuple().tm_yday)
```

```python
#     cal_month_no INT NOT NULL,
date_df['cal_month_number'] = date_df['date'].apply(lambda x: x.month)
```

```python
#     cal_month_name VARCHAR(20) NOT NULL,
date_df['cal_month_name'] = date_df['date'].apply(lambda x: x.strftime("%B"))
#     cal_month_abv VARCHAR(3) NOT NULL,
date_df['cal_month_abv'] = date_df['date'].apply(lambda x: x.strftime("%b"))
```

```python
#     cal_day_of_month INT NOT NULL,
date_df['day'] = date_df['date'].apply(lambda x: x.day)
```

```python
date_df.drop(columns=['date'], inplace=True)
```

```python
date_df.to_csv('dim_date.csv', index=False)
```
